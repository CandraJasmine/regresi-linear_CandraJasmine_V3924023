```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([10, 20, 30, 40, 50])
y1 = np.array([200, 400, 600, 800, 1000])
m1, b1 = 20, 0  # y = 20x

plt.scatter(x1, y1, color='blue')
plt.plot(x1, m1 * x1 + b1, color='red')
plt.title("No.1: Y = 20X")
plt.xlabel("Pelanggan")
plt.ylabel("Pendapatan (ribu)")
plt.grid(True)
plt.show()
```

```python
x2 = np.array([1, 2, 3, 4])
y2 = np.array([60, 65, 70, 74])
m2, b2 = 5, 55  # y = 5x + 55

plt.scatter(x2, y2, color='green')
plt.plot(x2, m2 * x2 + b2, color='red')
plt.title("No.2: Y = 5X + 55")
plt.xlabel("Jam Belajar")
plt.ylabel("Nilai")
plt.grid(True)
plt.show()
```

```python
x3 = np.array([1, 3, 5, 7, 9])
y3 = np.array([150, 120, 90, 60, 30])
m3, b3 = -15, 165  # y = -15x + 165

plt.scatter(x3, y3, color='purple')
plt.plot(x3, m3 * x3 + b3, color='red')
plt.title("No.3: Y = -15X + 165")
plt.xlabel("Umur Kendaraan (tahun)")
plt.ylabel("Nilai Kendaraan (juta)")
plt.grid(True)
plt.show()
```

```python
x4 = np.array([5, 10, 15, 20])
y4 = np.array([25000, 50000, 75000, 100000])
m4, b4 = 5000, 0  # y = 5000x

plt.scatter(x4, y4, color='orange')
plt.plot(x4, m4 * x4 + b4, color='red')
plt.title("No.4: Y = 5000X")
plt.xlabel("Jam Kerja")
plt.ylabel("Pendapatan (rupiah)")
plt.grid(True)
plt.show()
```

```python
x5_1 = np.array([50, 60, 70, 80, 90])
x5_2 = np.array([2, 3, 4, 5, 6])
y5 = np.array([400, 480, 560, 640, 720])
a5, b5_1, b5_2 = -320, 8, 80  # Y = -320 + 8X1 + 80X2

scatter = plt.scatter(x5_1, y5, c=x5_2, cmap='winter', s=100, edgecolor='black')
plt.colorbar(scatter, label='X2: Jumlah Komplain')
plt.title("No.5: Visualisasi Regresi Linier Berganda (2D)")
plt.xlabel("X1: Waktu Layanan")
plt.ylabel("Tingkat Kepuasan")
plt.grid(True)
plt.show()

explanation = """
Keterangan Visualisasi Soal No.5:
- Grafik menunjukkan hubungan antara dua variabel input (X1: waktu layanan, dan X2: jumlah komplain) terhadap tingkat kepuasan.
- Warna menunjukkan variasi nilai X2.
- Semakin tinggi waktu layanan dan semakin sedikit komplain, maka tingkat kepuasan meningkat.
"""
print(explanation)
```
